```python
import pandas as pd
import matplotlib.pyplot as plt

rumorose_df = pd.read_csv('Immagini_Rumorose/risultati_rumorose.csv')
restaurate_df = pd.read_csv('Immagini_Restaurate/risultati_restaurate.csv')

merged_df = pd.merge(rumorose_df, restaurate_df, on='ID', suffixes=('_rumorosa', '_restaurata'))

# funzione per visualizzare i grafici a barre per un singolo ID, divisi per metrica
def barplot_metrics_for_id(id_val):
    id_data = merged_df[merged_df['ID'] == id_val]
    if id_data.empty:
        print(f"ID {id_val} non trovato.")
        return

    metrics = ['PSNR', 'MSE', 'SSIM']
    ranges = {
        'PSNR': (10, 40 * 1.2),
        'MSE': (0, id_data[[f'{metric}_rumorosa' for metric in metrics]].max().max() * 1.2),
        'SSIM': (0, 1 * 1.2)
    }

    for metric in metrics:
        plt.figure(figsize=(10, 5))
        values = [id_data[f'{metric}_rumorosa'].values[0], id_data[f'{metric}_restaurata'].values[0]]
        bars = plt.bar(['Rumorosa', 'Restaurata'], values, color=['#b22222', '#32cd32'])

        for bar, value in zip(bars, values):
            plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05 * ranges[metric][1], f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='black', weight='bold')

        plt.xlabel('Tipo Immagine')
        plt.ylabel(f'Valore della Metrica {metric}')
        plt.ylim(ranges[metric])
        plt.title(f'{metric} - Immagine ID: {id_val}')
        plt.grid(axis='y')
        plt.show()

# funzione per visualizzare i grafici a linee per un singolo ID, divisi per metrica
def lineplot_metrics_for_id(id_val):
    id_data = merged_df[merged_df['ID'] == id_val]
    if id_data.empty:
        print(f"ID {id_val} non trovato.")
        return

    metrics = ['PSNR', 'MSE', 'SSIM']
    metrics_colors = {'PSNR': 'blue', 'MSE': 'red', 'SSIM': 'orange'}
    
    for metric in metrics:
        plt.figure(figsize=(10, 5))
        values = [id_data[f'{metric}_rumorosa'].values[0], id_data[f'{metric}_restaurata'].values[0]]
        plt.plot(['Rumorosa', 'Restaurata'], values, marker='o', label=metric, color=metrics_colors[metric])

        for i, value in enumerate(values):
            plt.text(i, value + (0.01 * value), f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='black', weight='bold')

        plt.xlabel('Tipo Immagine')
        plt.ylabel(f'Valore della Metrica {metric}')
        plt.title(f'{metric} - Immagine ID: {id_val}')
        plt.grid(True)
        plt.show()

# esempio con ID compreso tr 1 e 12
lineplot_metrics_for_id(9)
barplot_metrics_for_id(9)

print("\nRiassunto delle metriche per immagini rumorose e restaurate:")
print(merged_df.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

rumorose_df = pd.read_csv('Immagini_Rumorose/risultati_rumorose.csv')
restaurate_df = pd.read_csv('Immagini_Restaurate/risultati_restaurate.csv')

rumorose_df['BaseName'] = rumorose_df['NomeFile'].str.replace('_rumorosa', '').str.replace('.jpg', '')
restaurate_df['BaseName'] = restaurate_df['NomeFile'].str.replace('_restaurata', '').str.replace('.jpg', '')

merged_df = pd.merge(rumorose_df, restaurate_df, on='BaseName', suffixes=('_Rumorosa', '_Restaurata'))

merged_df['Delta_PSNR'] = merged_df['PSNR_Restaurata'] - merged_df['PSNR_Rumorosa']
merged_df['Delta_MSE'] = merged_df['MSE_Rumorosa'] - merged_df['MSE_Restaurata']
merged_df['Delta_SSIM'] = merged_df['SSIM_Restaurata'] - merged_df['SSIM_Rumorosa']

rumore_gaussiano = merged_df[merged_df['NomeFiltro_Rumorosa'] == 'Rumore Gaussiano']
rumore_sale_pepe = merged_df[merged_df['NomeFiltro_Rumorosa'] == 'Rumore Sale e Pepe']
rumore_periodico = merged_df[merged_df['NomeFiltro_Rumorosa'] == 'Rumore Periodico']

improvement_summary = {}

def calculate_percentage_improvement(group, metric):
    improvement = ((group[f'{metric}_Restaurata'] - group[f'{metric}_Rumorosa']) / group[f'{metric}_Rumorosa']).mean() * 100
    return improvement

noise_types = {
    'Rumore Gaussiano': rumore_gaussiano,
    'Rumore Sale e Pepe': rumore_sale_pepe,
    'Rumore Periodico': rumore_periodico
}

for noise, group in noise_types.items():
    improvement_summary[noise] = {
        'PSNR_Improvement (%)': calculate_percentage_improvement(group, 'PSNR'),
        'MSE_Improvement (%)': calculate_percentage_improvement(group, 'MSE'),
        'SSIM_Improvement (%)': calculate_percentage_improvement(group, 'SSIM')
    }

improvement_df = pd.DataFrame(improvement_summary).T
print(improvement_df)

metrics = ['PSNR_Improvement (%)', 'MSE_Improvement (%)', 'SSIM_Improvement (%)']
titles = ['PSNR Improvement (%)', 'MSE Improvement (%)', 'SSIM Improvement (%)']
colors = ['royalblue', 'indianred', 'gold']

for i, metric in enumerate(metrics):
    fig, ax = plt.subplots(figsize=(8, 6))
    improvement_df[metric].plot(kind='bar', ax=ax, color=colors[i])
    ax.set_title(titles[i])
    ax.set_ylabel('Percentage Improvement')
    ax.grid(True)
    ax.set_xticklabels([label.replace('Sale e Pepe', 'Sale e pepe').replace(' ', '\n', 1) for label in improvement_df.index], rotation=0, ha='center')
    plt.tight_layout()
    plt.show()
```

```python
average_improvement = improvement_df.mean()
print("\nAverage Improvement for Each Metric:")
print(average_improvement)
```
