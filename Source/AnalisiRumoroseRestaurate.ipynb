```python
import pandas as pd
import matplotlib.pyplot as plt

# Caricare i CSV
rumorose_df = pd.read_csv('Immagini_Rumorose/risultati_rumorose.csv')
restaurate_df = pd.read_csv('Immagini_Restaurate/risultati_restaurate.csv')

# Unire i DataFrame per ID
merged_df = pd.merge(rumorose_df, restaurate_df, on='ID', suffixes=('_rumorosa', '_restaurata'))

# Funzione per visualizzare i grafici a barre per un singolo ID, divisi per metrica
def barplot_metrics_for_id(id_val):
    id_data = merged_df[merged_df['ID'] == id_val]
    if id_data.empty:
        print(f"ID {id_val} non trovato.")
        return

    metrics = ['PSNR', 'MSE', 'SSIM']
    ranges = {
        'PSNR': (10, 40),
        'MSE': (0, id_data[[f'{metric}_rumorosa' for metric in metrics]].max().max() + 10),
        'SSIM': (0, 1)
    }

    for metric in metrics:
        plt.figure(figsize=(10, 5))
        plt.bar(['Rumorosa', 'Restaurata'], [id_data[f'{metric}_rumorosa'].values[0], id_data[f'{metric}_restaurata'].values[0]], color=['red', 'green'])
        plt.xlabel('Tipo Immagine')
        plt.ylabel(f'Valore della Metrica {metric}')
        plt.ylim(ranges[metric])
        plt.title(f'Confronto {metric} tra Immagini Rumorose e Restaurate per ID {id_val}')
        plt.grid(axis='y')
        plt.show()

# Funzione per visualizzare i grafici a linee per un singolo ID, divisi per metrica
def lineplot_metrics_for_id(id_val):
    id_data = merged_df[merged_df['ID'] == id_val]
    if id_data.empty:
        print(f"ID {id_val} non trovato.")
        return

    metrics = ['PSNR', 'MSE', 'SSIM']
    for metric in metrics:
        plt.figure(figsize=(10, 5))
        plt.plot(['Rumorosa', 'Restaurata'], [id_data[f'{metric}_rumorosa'].values[0], id_data[f'{metric}_restaurata'].values[0]], marker='o', label=metric, color='blue')
        plt.xlabel('Tipo Immagine')
        plt.ylabel(f'Valore della Metrica {metric}')
        plt.title(f'Confronto {metric} tra Immagini Rumorose e Restaurate per ID {id_val}')
        plt.grid(True)
        plt.show()

# Esempio di utilizzo delle funzioni
lineplot_metrics_for_id(1)
barplot_metrics_for_id(1)

# Riassunto Statistico
print("\nRiassunto delle metriche per immagini rumorose e restaurate:")
print(merged_df.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Caricare i CSV
rumorose_df = pd.read_csv('Immagini_Rumorose/risultati_rumorose.csv')
restaurate_df = pd.read_csv('Immagini_Restaurate/risultati_restaurate.csv')

# Unire i DataFrame per ID
merged_df = pd.merge(rumorose_df, restaurate_df, on='ID', suffixes=('_rumorosa', '_restaurata'))

# Funzione per visualizzare i grafici a barre per un singolo ID, divisi per metrica
def barplot_metrics_for_id(id_val):
    id_data = merged_df[merged_df['ID'] == id_val]
    if id_data.empty:
        print(f"ID {id_val} non trovato.")
        return

    metrics = ['PSNR', 'MSE', 'SSIM']
    ranges = {
        'PSNR': (10, 40 * 1.2), # Estendere leggermente il limite per evitare che PSNR esca dalla griglia
        'MSE': (0, id_data[[f'{metric}_rumorosa' for metric in metrics]].max().max() * 1.2),  # Estendere leggermente il limite per evitare che MSE esca dalla griglia
        'SSIM': (0, 1 * 1.2)  # Estendere leggermente l'asse y per evitare che SSIM esca dalla griglia
    }

    for metric in metrics:
        plt.figure(figsize=(10, 5))
        values = [id_data[f'{metric}_rumorosa'].values[0], id_data[f'{metric}_restaurata'].values[0]]
        bars = plt.bar(['Rumorosa', 'Restaurata'], values, color=['red', 'green'])

        # Aggiungere i valori sulle barre, leggermente sopra la barra
        for bar, value in zip(bars, values):
            plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05 * ranges[metric][1], f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='black', weight='bold')

        plt.xlabel('Tipo Immagine')
        plt.ylabel(f'Valore della Metrica {metric}')
        plt.ylim(ranges[metric])
        plt.title(f'Confronto {metric} tra Immagini Rumorose e Restaurate per ID {id_val}')
        plt.grid(axis='y')
        plt.show()

# Funzione per visualizzare i grafici a linee per un singolo ID, divisi per metrica
def lineplot_metrics_for_id(id_val):
    id_data = merged_df[merged_df['ID'] == id_val]
    if id_data.empty:
        print(f"ID {id_val} non trovato.")
        return

    metrics = ['PSNR', 'MSE', 'SSIM']
    for metric in metrics:
        plt.figure(figsize=(10, 5))
        values = [id_data[f'{metric}_rumorosa'].values[0], id_data[f'{metric}_restaurata'].values[0]]
        plt.plot(['Rumorosa', 'Restaurata'], values, marker='o', label=metric, color='blue')

        # Aggiungere i valori sulle linee, leggermente sopra il punto per evitare sovrapposizioni
        for i, value in enumerate(values):
            plt.text(i, value + (0.01 * value), f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='black', weight='bold')

        plt.xlabel('Tipo Immagine')
        plt.ylabel(f'Valore della Metrica {metric}')
        plt.title(f'Confronto {metric} tra Immagini Rumorose e Restaurate per ID {id_val}')
        plt.grid(True)
        plt.show()

# Esempio di utilizzo delle funzioni
lineplot_metrics_for_id(1)
barplot_metrics_for_id(1)

# Riassunto Statistico
print("\nRiassunto delle metriche per immagini rumorose e restaurate:")
print(merged_df.describe())
```
