```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV files
rumorose_df = pd.read_csv('Immagini_Rumorose/risultati_rumorose.csv')
restaurate_df = pd.read_csv('Immagini_Restaurate/risultati_restaurate.csv')

# Converting 'ID' column to categorical for easier comparison between noisy and restored images
rumorose_df['ID'] = rumorose_df['ID'].astype(str)
restaurate_df['ID'] = restaurate_df['ID'].astype(str)

# Merging the dataframes on 'ID' to directly compare noisy and restored metrics
merged_df = pd.merge(rumorose_df, restaurate_df, on='ID', suffixes=('_rumorosa', '_restaurata'))

# Plotting comparison metrics PSNR, MSE, SSIM between noisy and restored images
plt.figure(figsize=(18, 12))

# PSNR comparison
plt.subplot(3, 1, 1)
sns.boxplot(data=[merged_df['PSNR_rumorosa'], merged_df['PSNR_restaurata']], palette="Set2")
plt.xticks([0, 1], ['Rumorosa', 'Restaurata'])
plt.title('PSNR - Immagini Rumorose vs Restaurate')
plt.ylabel('PSNR')

# MSE comparison
plt.subplot(3, 1, 2)
sns.boxplot(data=[merged_df['MSE_rumorosa'], merged_df['MSE_restaurata']], palette="Set2")
plt.xticks([0, 1], ['Rumorosa', 'Restaurata'])
plt.title('MSE - Immagini Rumorose vs Restaurate')
plt.ylabel('MSE')

# SSIM comparison
plt.subplot(3, 1, 3)
sns.boxplot(data=[merged_df['SSIM_rumorosa'], merged_df['SSIM_restaurata']], palette="Set2")
plt.xticks([0, 1], ['Rumorosa', 'Restaurata'])
plt.title('SSIM - Immagini Rumorose vs Restaurate')
plt.ylabel('SSIM')

plt.tight_layout()
plt.show()

# Calculating improvement statistics for each filter
improvement_stats = merged_df.copy()
improvement_stats['PSNR_Improvement'] = improvement_stats['PSNR_restaurata'] - improvement_stats['PSNR_rumorosa']
improvement_stats['MSE_Improvement'] = improvement_stats['MSE_rumorosa'] - improvement_stats['MSE_restaurata']
improvement_stats['SSIM_Improvement'] = improvement_stats['SSIM_restaurata'] - improvement_stats['SSIM_rumorosa']

# Aggregating improvement statistics by filter
filter_stats = improvement_stats.groupby('NomeFiltro_restaurata')[['PSNR_Improvement', 'MSE_Improvement', 'SSIM_Improvement']].mean()

# Displaying the improvement statistics dataframe
display(filter_stats)  # Use display if running in Jupyter Notebook
```
